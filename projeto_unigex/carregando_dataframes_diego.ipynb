Carregando e salvado os DataFrames

```python
import pandas as pd
import os
from joblib import dump

# Criação da pasta
os.makedirs('util/data', exist_ok=True)

# Categorias
# 1 - Carregando o arquivo CSV de categorias
df_categorias = pd.read_csv("categorias.csv")

# 2 - Salvando o DataFrame utilizando joblib
dump(df_categorias,'util/data/df_categorias.z')
print("DataFrame salvo em util/data com sucesso!")

# Produtos
# 1 - Carregando o arquivo CSV de produtos
df_produtos = pd.read_csv("produtos.csv")

# 2 - Salvando o DataFrame utilizando joblib
dump(df_produtos,'util/data/df_produtos.z')
print("DataFrame salvo em util/data com sucesso!")

# Unidades
# 1 - Carregando o arquivo CSV de unidades
df_unidades = pd.read_csv("unidades.csv")

# 2 - Salvando o DataFrame utilizando joblib
dump(df_unidades,'util/data/df_unidades.z')
print("DataFrame salvo em util/data com sucesso!")

# Vendas
# 1 - Carregando o arquivo CSV de vendas
df_vendas = pd.read_csv("vendas.csv")

# 2 - Salvando o DataFrame utilizando joblib
dump(df_vendas,'util/data/df_vendas.z')
print("DataFrame salvo em util/data com sucesso!")

# Exibição dos DataFrame:

```
